```python






#classifing cupcake and muffin with classification tecnique

########### package for analysis
import pandas as pd
import numpy as np
from sklearn import svm

####packaages for data visualising

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2)
```

```python

recipes=pd.read_csv('recipes_muffins_cupcakes_scones.csv')
recipes
```

```python
ingridients1=recipes[['Flour','Milk','Sugar','Butter','Egg','Baking Powder','Vanilla','Salt']]
```

```python
ingridients1
```

```python
data1=np.array(ingridients1)
```

```python
data1
```

```python
ingridients1.plot()
```

```python
label=[]
for x in recipes['Type']:
    if x=='Muffin':
        label.append(0)
    elif x=='Cupcake':
        label.append(1)
    else:
        label.append(2)
label_data=np.array(label)
```

```python
label_data
```

```python
model=svm.SVC(kernel='linear')
model.fit(ingridients1,label_data)
```

```python
def muffin_or_cupcake_or_Scone(flour,Milk,Sugar,Butter,Egg,Baking_powder,Vanila,Salt):
    status=model.predict([[flour,Milk,Sugar,Butter,Egg,Baking_powder,Vanila,Salt]])
    if status==0:
        print('You\'re looking at a muffin recipe!')
    elif status==1:
        print('You\'re looking at a cupcake recipe!')
    else:
        print('You\'re looking at a Scone recipe!')
```

```python
muffin_or_cupcake_or_Scone(19,17,51,5,0,10,0,5)
```

```python
muffin_or_cupcake_or_Scone(45, 11, 17, 17,  8,  1,  0,  0)
```
